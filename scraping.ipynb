```python
# 1. Install library terlebih dahulu (jalankan di cell Colab)
# !pip install youtube-comment-downloader pandas

from youtube_comment_downloader import *
import pandas as pd
import itertools

# 2. Inisialisasi downloader
downloader = YoutubeCommentDownloader()

# 3. Masukkan link video YouTube yang dituju
# Pastikan video ini punya lebih dari 10.000 komentar
video_url = 'https://youtu.be/VQRLujxTm3c?si=SDGPZG-1UGd8rr2n'

print("Mulai mengunduh komentar, proses ini mungkin memakan waktu beberapa menit...")

# 4. Mengambil komentar (diurutkan dari yang terpopuler atau terbaru)
# SORT_BY_POPULAR atau SORT_BY_RECENT
comments = downloader.get_comments_from_url(video_url, sort_by=SORT_BY_POPULAR)

# 5. Membatasi jumlah pengambilan data agar mencapai target Bintang 5 (10.000 data)
# Kita lebihkan sedikit (misal 10.500) untuk cadangan jika ada komentar kosong/spam
limit = 12000
comments_list = list(itertools.islice(comments, limit))

# 6. Mengubah data menjadi bentuk Tabel (DataFrame)
df = pd.DataFrame(comments_list)

# 7. Memfilter hanya kolom teks komentar saja yang kita ambil
# Kamu juga bisa mengambil kolom 'author' atau 'time' jika butuh
df_clean = df[['text']]

# Membersihkan baris yang kosong (jika ada)
df_clean = df_clean.dropna()

# 8. Menyimpan data hasil scraping ke dalam format CSV
file_name = 'dataset_komentar_youtube.csv'
df_clean.to_csv(file_name, index=False)

print(f"Selesai! Berhasil menyimpan {len(df_clean)} komentar ke dalam file {file_name}.")
```
